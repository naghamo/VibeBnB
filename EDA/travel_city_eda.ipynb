```python
from pyspark.sql import functions as F
from pyspark.sql import SparkSession, DataFrame
from config import *
from airbnb_data_loader import load_airbnb_data

def rows_per_continent(airbnb_df, continents: dict):
    

    return (
        airbnb_df
        .select("addr_cc")
        .withColumn("continent", map_expr.getItem(F.col("addr_cc")))
        .fillna({"continent": "other"})
        .groupBy("continent")
        .agg(F.count("*").alias("n_rows"), F.countDistinct("addr_cc").alias("n_countries"))
        .orderBy(F.col("n_rows").desc())
    )

# usage

display(rows_per_continent(airbnb_df, continents))
```
